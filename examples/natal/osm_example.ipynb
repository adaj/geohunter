```python
PACKAGE_FOLDER = '/home/adelsondias/Repos/geohunter'

import sys
sys.path.append(PACKAGE_FOLDER)
import geohunter
```

```python
import geopandas as gpd

districts = gpd.read_file('bairros_natal.geojson')
districts.plot()
```

```python
bbox =  {'north':districts.bounds.max().values[3],
         'east':districts.bounds.max().values[2],
         'south':districts.bounds.min().values[1],
         'west':districts.bounds.min().values[0]}

hawk = geohunter.osm.OSMHawk(bbox, folder='/home/adelsondias/Repos/geohunter/examples/natal/')
```

```python
pdf, ldf = hawk.transform(points={
    'amenity':['hospital', 'place_of_worship', 'police'],
    'leisure':['*'],
    'tourism':['*'],
    'shop':['*'],
    'highway':['bus_stop']
}, lines={
    'highway':['residential', 'primary']
})
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(4,6))
pdf.plot(column='tag', ax=ax)
```
